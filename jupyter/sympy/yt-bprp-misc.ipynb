```python
import logging

import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
from latex_utils import set_display_level, sym_display_latex

set_display_level(logging.INFO)
```

### [How to solve an exponential function with quadratic exponent? Reddit r/homeworkhelp](https://youtu.be/NTyrJW8a_98)

```python
x = sym.symbols('x')
lhs = 16 ** (x**2 + x)
sym_display_latex('lhs=', lhs)

rhs = sym.Integer(1) / 2

eq = sym.Eq(lhs, rhs)
sym_display_latex('', eq, display_level=logging.INFO)

solset = sym.solveset(eq, x, domain=sym.S.Reals)
sym_display_latex('solset=', solset, display_level=logging.INFO)
```

```python
xmin = -2
xmax = 1

# draw guide lines
ax_xlim = [xmin, xmax]
ax_ylim = [-10, 20]
plt.plot(ax_xlim, [0, 0], '--', color='lightgray')
plt.plot([0, 0], ax_ylim, '--', color='lightgray')

xvals = np.linspace(xmin, xmax, 201)
sym_display_latex('xvals=', xvals)
yvals = 16 ** (xvals**2 + xvals) - (1/2)
sym_display_latex('yvals=', yvals)

plt.plot(xvals, yvals, '--', label=f'$f(x) = {sym.latex(lhs - rhs)}$')
# plt.plot([xvals[[0]]], [yvals[[0]]], 'g.')

xvals_root = np.where(xvals == -(1/2))
xvals_root = xvals_root[0] if len(xvals_root) > 0 else xvals_root
sym_display_latex('xvals_root=', xvals_root)

plt.plot(xvals[xvals_root], yvals[xvals_root], 'rs', label=f'root = ({xvals[xvals_root][0]}, {yvals[xvals_root][0]})')
plt.text(
    xvals[xvals_root] - 0.2,
    yvals[xvals_root] + 1,
    f'root = ({xvals[xvals_root][0]}, {yvals[xvals_root][0]})',
    fontsize=8
)

# plt.axis('square')
plt.axis([*ax_xlim, *ax_ylim])
plt.legend()
plt.xlabel('x')
plt.ylabel(f'$f(x) = {sym.latex(lhs - rhs)}$')
plt.title(f'The single Real root of ${sym.latex(eq)}$.')
plt.show()
```
